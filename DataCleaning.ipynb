### Ex. 1 Have a look at the following dataset: dsm-beuth-edl-demodata-orig

Write a Python / Panda Script which 'cleans' this data set. Justify your actions in the respective notebook or python script you provide as a solution (link, file, kaggle repo, etc.).

The original dataset does not necessarily have to be created. A proper strategy / good arguments are more important.

```python
import pandas as pd
import numpy as np
url="https://raw.githubusercontent.com/edlich/eternalrepo/master/DS-WAHLFACH/dsm-beuth-edl-demodata-dirty.csv"
data = pd.read_csv(url, sep=",")
df = pd.DataFrame(data)
df
```

#### First of all, get rid of completely empty rows.

```python
df.dropna(subset = ["full_name"], inplace=True)

df
```

#### Delete duplicates.

```python
df = df.drop_duplicates(subset=['full_name', 'first_name', 'last_name', 'email', 'gender', 'age'], keep='last')
df
```

#### Give an ID where ID is NaN.

```python
df['id'] = df['id'].replace(np.nan, np.max(df['id']) + 1)
df
```

#### Replace NaN in email column.

```python
df['email'] = df['email'].replace(np.nan, 'unknown')
df['gender'] = df['gender'].replace(np.nan, 'unknown')

df
```

#### Replace the word 'old' in the age column with a 0.
(I don't know what they would do with it, but since there is no way to find out the real age and we want to do data science on this data, this might be a good compromise.)

```python
df['age'] = df['age'].replace(regex = '([a-zA-Z])', value = 0)
df
```

#### Convert the age column entries into numerics.

```python
df['age'] = pd.to_numeric(df['age'])
df
```

#### Convert negative values in age column to positive values.

```python
df.loc[df.age < 0, 'age'] = (df.age * -1)
df
```
